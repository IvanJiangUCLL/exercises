```python
import pandas as pd
import numpy as np
import ast
import matplotlib.pyplot as plt
import seaborn as sns

from datetime import datetime
from dateutil.parser import parse
```

```python
filepath = "../netflix.csv"
data = pd.read_csv(filepath, index_col=0)
```

```python
data.head()
```

```python
data.shape
data.info()
```

```python
total_null = data.isnull().sum().sort_values(ascending = False)
percent = ((data.isnull().sum()/data.isnull().count())*100).sort_values(ascending = False)
print("Total records = ", data.shape[0])

missing_data = pd.concat([total_null, percent.round(2)], axis=1, keys=["Total Missing", "In Percent"])
missing_data
```

```python
data["Rating"] = data["Rating"].replace(np.nan, "Unknown Rating")
data["Director"] = data["Director"].replace(np.nan, "Unknown Director")
data["Cast"] = data["Cast"].replace(np.nan, "[Unknown Cast]")
data["Country"] = data["Country"].replace(np.nan, "Unknown Country")
```

```python
total_null = data.isnull().sum().sort_values(ascending = False)
percent = ((data.isnull().sum()/data.isnull().count())*100).sort_values(ascending = False)
print("Total records = ", data.shape[0])

missing_data = pd.concat([total_null,percent.round(2)],axis=1,keys=['Total Missing','In Percent'])
missing_data
```

```python
duplicates = data[data.duplicated()]
duplicates.head()
```

```python
data = data.drop_duplicates()
data.shape
```

```python
dt = datetime.now()

dt_str = dt.strftime("%b %d, %Y")

data["Release_Date"] = data["Release_Date"].replace(np.nan, dt_str )

data["Release_Date"] = pd.to_datetime(data["Release_Date"])


data["Day"] = data["Release_Date"].dt.day
data["Month"] = data["Release_Date"].dt.strftime("%b")
data["Year"] = data["Release_Date"].dt.year


data["Week day"] = data["Release_Date"].dt.day_name()
```
